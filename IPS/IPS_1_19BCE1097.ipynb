Import an inbuilt dataframe from the library

```python
boston = load_boston()
display(boston)
```

```python
boston_df = pd.DataFrame(boston.data, columns = boston.feature_names)
boston_df
```

```python
boston_df['MEDV'] = boston.target
```

```python
display(boston_df)
```

```python
# Number of Not missing cell/rows
boston_df.isnull().sum()
```

```python
# Number of Not missing cell/rows
boston_df.notnull().sum()
```

# Line Plotting
*   Use line plot for univariante data analysis(continous)



```python
plt.figure(figsize=(18,8))
plt.plot(boston_df['AGE'])
plt.title("Age vs Rows Line Plot", fontsize = 10)
plt.xlabel("Row Indexes")
plt.ylabel("AGE")
plt.show() 
```

#BAR Plot
*   Bivariate Visualization - X must be discrete & Y can be discrete or continious




```python
plt.figure(figsize=(15,9))
plt.bar(x = boston_df['ZN'], height = boston_df['AGE'], width = 1.5, color = 'deeppink')
plt.title("Age vs ZN", fontsize = 10)
plt.xlabel("ZN")
plt.ylabel("AGE")
plt.show() 
```

#Histogram
*   Frequency of Data in an interval - variance in Data
*   Univariate Analysis of Continious 

```python
plt.figure(figsize=(13,9))
plt.hist(boston_df['MEDV'], bins = 15, color = 'darkviolet')
plt.title("Histogram Plot", fontsize = 10)
plt.xlabel("MEDV Interval")
plt.ylabel("Frequency in Range")
plt.show() 
```

```python
iris_df = pd.read_csv("/content/drive/MyDrive/Datasets/Iris.csv")
iris_df
```

```python
iris_df['Species'].unique() 
```

```python
plt.figure(figsize=(13,8))
plt.scatter(x = iris_df['SepalLengthCm'], y = iris_df['PetalLengthCm'], c = 'blue')
plt.title("Scatter Plot", fontsize = 10)
plt.xlabel("Sepal Length")
plt.ylabel("Petal Length")
plt.show()
```

```python
plt.style.use('ggplot')
plt.figure(figsize=(13,8))
plt.scatter(x = iris_df['SepalLengthCm'], y = iris_df['PetalLengthCm'], c = 'r', marker='^', label = 'PetalLengthCm')
plt.scatter(x = iris_df['SepalLengthCm'], y = iris_df['PetalWidthCm'], c = 'b', marker = 'o', label = 'PetalWidthCm')
plt.title("Scatter Plot", fontsize = 10)
plt.xlabel("Sepal Length")
plt.ylabel("Petal Length/Petal Width")
plt.legend()
plt.show() 
```
