### **Simple Linear regression with Age as input attribute**

```python
from sklearn.linear_model import LinearRegression
```

```python
#Simple Linear Regression
X_simple_lreg = new_toyota_df[["Age"]].values
y_simple_lreg = new_toyota_df["Price"].values
print(X_simple_lreg[0:5])
print(y_simple_lreg[0:5])
```

```python
print((X_simple_lreg).ndim)
print((y_simple_lreg).ndim)
```

```python
#Create train validation dataset
from sklearn.model_selection import train_test_split
X_train_slreg, X_val_slreg, y_train_slreg, y_val_slreg = train_test_split(X_simple_lreg,y_simple_lreg, test_size = 0.25, random_state = 4)
print('Train Dataset : ', X_train_slreg.shape, y_train_slreg.shape)
print('Validation Dataset : ', X_val_slreg.shape, y_val_slreg.shape)
```

```python
simple_lreg = LinearRegression()
simple_lreg.fit(X_train_slreg, y_train_slreg)
print('Intercept : ', simple_lreg.intercept_)
print('Slope : ', simple_lreg.coef_)
```

```python
# Use the model to predict the validation dataset.
y_simplelreg_pred_val = simple_lreg.predict(X_val_slreg)
```

```python
# Use the model to predict the train dataset.
y_simplelreg_pred_train = simple_lreg.predict(X_train_slreg)
```

```python
# Calculate the evaluation metrics of the model.
mse_slreg = (mean_squared_error(y_simplelreg_pred_val, y_val_slreg))
print('Mean squared error for simple linear reg : ', mse_slreg)
```

```python
img = cv2.imread('/content/drive/MyDrive/Datasets/Week 4/r-squared.png', cv2.IMREAD_UNCHANGED)
cv2_imshow(img)
```

```python
from sklearn.metrics import r2_score
r2_score_slreg_train = r2_score(y_simplelreg_pred_train, y_train_slreg)
r2_score_slreg_val = r2_score(y_simplelreg_pred_val, y_val_slreg)
print('r2_ score for train dataset for simple linear reg : ', r2_score_slreg_train)
print('r2_ score for test dataset for simple linear reg : ', r2_score_slreg_val)
```

# **Multiple Regression using all input attributes**

```python
# Separating the independent and dependent variable.
X_multi_lreg = data_df.drop('Price', axis = 1).values
y_multi_lreg = data_df["Price"].values.reshape(-1,1)
```

```python
X_multi_lreg
```

```python
y_multi_lreg
```

```python
# Create train and validation  dataset
from sklearn.model_selection import train_test_split
X_train_mlreg, X_val_mlreg, y_train_mlreg, y_val_mlreg = train_test_split(X_multi_lreg,y_multi_lreg, test_size = 0.25, random_state = 4)
print('Train Dataset : ', X_train_mlreg.shape, y_train_mlreg.shape)
print('Test Dataset : ', X_val_mlreg.shape, y_val_mlreg.shape)
```

```python
multi_lreg = LinearRegression()
multi_lreg.fit(X_train_mlreg, y_train_mlreg)
print('Intercept : ', multi_lreg.intercept_)
print('Slope : ', multi_lreg.coef_)
```

```python
# Use the model to predict the test dataset.
y_mlreg_pred_val = multi_lreg.predict(X_val_mlreg)
```

```python
# Use the model to predict the train dataset.
y_mlreg_pred_train = multi_lreg.predict(X_train_mlreg)
```

```python
# Calculate the evaluation metrics of the model.
rmse_mlreg = np.sqrt(mean_squared_error(y_mlreg_pred_val, y_val_mlreg)**2)
print('root mean squared error for multi linear reg : ', rmse_mlreg)
```

```python
from sklearn.metrics import r2_score
r2_score_mlreg_train = r2_score(y_mlreg_pred_train, y_train_mlreg)
r2_score_mlreg_val = r2_score(y_mlreg_pred_val, y_val_mlreg)
print('r2_ score for train dataset for multi linear reg : ', r2_score_mlreg_train)
print('r2_ score for test dataset for multi linear reg : ', r2_score_mlreg_val)
```
