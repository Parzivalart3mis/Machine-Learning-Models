# **9. Create a dataframe of a dataset 'df'. Slice a portion of 'df' which is in the shape of a square, whose size equals the number of columns of 'df'. If 'dfslice' contains non-numerical items, replace those entries by zero. Considering the slice 'dfslice' as a square matrix, compute the inverse of 'dfslice' and output the `inversedslice'.**

```python
df =  pd.read_csv('/content/drive/MyDrive/Datasets/anime.csv')
df
```

```python
dfslice = df.iloc[0:7,:7]
dfslice
```

```python
for col in dfslice.columns:    
    dfslice[col] =pd.to_numeric(dfslice[col], errors ='coerce').fillna(0).astype('float64')
dfslice
```

```python
inversedslice = pd.DataFrame(np.linalg.pinv(dfslice.values), dfslice.columns, dfslice.index)
inversedslice
```
